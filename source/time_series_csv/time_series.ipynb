```python
import pandas as pd
import glob
```

```python
timeseries = []

for year in range(2009,2019):
    filename = 'Registro ' + str(year) + '.csv'
    df = pd.read_csv(filename,sep=';',skiprows=1)
    print(f'{year} - Before cleaning: {df.shape}')
    # Columns
    df['ano'] = year
    # df['UF_old'] = df['UF']
    # Fill down
    df['UF'] = df['UF'].fillna(method='pad')
    # Remove totals
    df = df[(df.UF != 'TODAS') & (df.CATEGORIA.str.contains('TOTAL') == False)]
    print(f'{year} - After cleaning: {df.shape}')
    timeseries.append(df)
```

```python
df = pd.concat(timeseries, axis=0, ignore_index=True)
```

```python
df.to_csv('2009-2018.csv')
```
