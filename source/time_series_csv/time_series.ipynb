```python
import pandas as pd
```

```python
timeseries = []

for year in range(2009,2019):
    filename = 'Registro ' + str(year) + '.csv'
    df = pd.read_csv(filename,sep=';',skiprows=1)
    print(f'{year} - Before cleaning: {df.shape}')
    # Fill down
    # df['UF_old'] = df['UF']
    df['UF'] = df['UF'].fillna(method='pad')
    # Remove totals
    df = df[(df.UF != 'TODAS') & (df.CATEGORIA.str.contains('TOTAL|Total') == False)]   
    df['ano'] = year
    print(f'{year} - After cleaning: {df.shape}')
    timeseries.append(df)
```

```python
df = pd.concat(timeseries, axis=0, ignore_index=True)
```

```python
df.to_csv('2009-2018.csv',index=False)
```

# 2019 : changes in the 'category' variable

```python
df1_19 = pd.read_csv('Registro 2019 Jan Abr.csv',sep=';',skiprows=1)
df2_19 = pd.read_csv('Registro 2019 Abr Dez.csv',sep=';',skiprows=1)   

# Cleaning
df1_19['UF'] = df1_19['UF'].fillna(method='pad')
df1_19 = df1_19[(df1_19.UF != 'TODAS') & (df1_19.CATEGORIA.str.contains('TOTAL|Total') == False)]   

# df2
df2_19['UF'] = df2_19['UF'].fillna(method='pad')
df2_19 = df2_19[(df2_19.UF != 'TODAS') & (df2_19.CATEGORIA.str.contains('TOTAL|Total') == False)]  
```

```python
# Remove empty months (April - December)
# df2
print(df1_19.columns)
df1_19 = df1_19.iloc[:,:5]
print(df1_19.columns)
```

```python
# Old categories
df1_19.CATEGORIA.unique()
```

```python
# New categories
df2_19.CATEGORIA.unique()
```

```python
# Standardize categories

df2_19['NOVA_CATEGORIA'] = df2_19.CATEGORIA

df2_19.CATEGORIA = df2_19.CATEGORIA.replace({
    'Caçador de Subsistência': 'Pessoa física',
    'Cidadão': 'Pessoa física',
    'Servidor Público (Porte por prerrogativa de função)': 'Segurança Pública',
    'Empresa com Segurança Orgânica': 'Segurança Privada',
    'Empresa Comercial': 'Segurança Privada',
    'Empresa de Segurança Privada': 'Segurança Privada',
    'Fabricante/Importador': 'Loja de arma',
    'Órgão público com taxa': 'Órgão Público',
    'Órgão público sem taxa': 'Órgão Público',
    'Outras Categorias': 'Outros Tipos',
    'Revendedor': 'Loja de arma'
})
```

### Checking categories

```python
df1_19.CATEGORIA.unique()
```

```python
df2_19.CATEGORIA.unique()
```

```python
# df2 - group by old categories and delete last column
df2_19 = df2_19.groupby(['UF','CATEGORIA'], as_index=False).sum().iloc[:,:-1] 
```

```python
# delete columns 
cols_deleted = ['JANEIRO','FEVEREIRO ','MARÇO']

df2_19 = df2_19[df2_19.columns.difference(cols_deleted,sort=False)]
```

```python
df2_19
```

### Merge 2019

```python
print(f'Columns Jan- Mar >>>>>>>> {df1_19.columns}')
print(f'Shape Jan- Mar >>>>>>>> {df1_19.shape}')
```

```python
print(f'Columns Apr- Dez >>>>>>>> {df2_19.columns}')
print(f'Shape Apr- Dez >>>>>>>> {df2_19.shape}')
```

```python
keycolumns = ['UF','CATEGORIA']

df_19 = pd.merge(df1_19,df2_19,left_on=key_columns,right_on=keycolumns)
```

```python
# Total

df_19['TOTAL'] = df_19.loc[:, 'JANEIRO':'DEZEMBRO'].sum(axis=1)

df_19.to_csv('2019.csv',index=False)

df_19
```
