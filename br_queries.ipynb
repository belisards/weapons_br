```python
import pandas as pd
```

```python
df= pd.read_csv('time_series.csv')

df
```

```python
total = df.groupby(['ANO','NOVA_CATEGORIA'],as_index=False).sum()
```

```python
total.tail()
```

```python
total.ANO = total.ANO.astype('int')
total.TOTAL = total.TOTAL.astype('int')

columns = ['ANO','NOVA_CATEGORIA','TOTAL']

total[columns] 

```

```python
total.pivot_table(index=["NOVA_CATEGORIA"], 
                    columns='ANO', 
                    values='TOTAL').to_csv('total.csv')
```
