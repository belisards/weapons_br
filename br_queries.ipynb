```python
import pandas as pd
```

```python
df= pd.read_csv('time_series.csv')

df
```

```python
total = df.groupby(['ANO','CATEGORIA'],as_index=False).sum()
```

```python
total.ANO = total.ANO.astype('int')
total.TOTAL = total.TOTAL.astype('int')

columns = ['ANO','CATEGORIA','TOTAL']

total[columns] 

```

```python
total.pivot_table(index=["CATEGORIA"], 
                    columns='ANO', 
                    values='TOTAL').to_csv('total.csv')
```
